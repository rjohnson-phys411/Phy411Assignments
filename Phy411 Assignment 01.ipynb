**1** To start, make a time vector that goes from 0 to 100 s, with 10000 entries in it.  Then make a synthetic time series that has two close in frequency sine waves:

$$ x = A_1 \cos (\omega t) + A_2 \cos ((\omega+\delta \omega ) t) $$

where $\omega$ is the "carrier" frequency, and $\delta \omega << \omega$ is the difference in frequencies between the two waves.  For simplicty, you could make $A_1=A_2=1.0\ [V]$.  Feel free to choose an interesting $\omega$.

<hr>

### A1.1:

Assuming that $A_1=A_2=1.0\ [V]$ then:
$$x=\cos(\omega t)+\cos((\omega+\delta\omega)t)$$

```python
t=np.arange(0.,100.,0.1)
print 'This time vector goes from', t[1],' to ', t[100],' and has a ', np.shape(t)[0], ' entries.'
```

**2** Now plot the two component waves and the full value of $x$ as a time series.  Preferably on the same axis.  Make sure to label your axes and include a legend<hr>

```python
def mathstuff(Omega,DeltaOmega,T): 
    """A Function that finds the components with the given values for omega and delta omega,
    and a given time series (T). Returns an array of time, component wave 1, 2, 
    and full value of x"""
    x1 = np.cos(Omega*T)
    x2 = np.cos((Omega+DeltaOmega)*T) 
    x = x1+x2
    return np.array([T,x1,x2,x])

#Defining values of omega and delta omega.
omega=np.pi
deltaomega= np.array([0.01*np.pi])

#Calling mathstuff to determine component wave 1, 2, and full value of x with the time series.
#Returns np.array([T,x1,x2,x]).
Datastuff=mathstuff(omega,deltaomega[0],t)

#Plotting component wave 1, 2, and full value of x vs. the time series.
fig=plt.figure(figsize=(12,4))
ax=fig.add_subplot(1,1,1) 
ax.plot(Datastuff[0],Datastuff[1],label='$\cos(\pi t)$',color='b')
ax.plot(Datastuff[0],Datastuff[2],label='$\cos((\pi+0.01\pi)t)$',color='g')
ax.plot(Datastuff[0],Datastuff[3],label='$\cos(\pi t)+\cos((\pi+0.01\pi)t)$',color='k')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Measured Volts [V]') #lets say we're measuring the volt of something.
ax.set_title('Measured Volts vs. Time')
ax.set_ylim(top=3.)
ax.legend(loc=1)


```

**3** Write a function using the `np.random.norm` function to simulate the coin flip.  It shoudl return $+1/2$ twice as often as $-1/2$.  Demonstrate that your function works.
<hr>

```python
def HeadsOrTails(NUM_FLIPS,passout): 
    """Function determines if coin flip is heads or tails using np.random.rand. """
    passover=np.random.rand(1,NUM_FLIPS)[0]
    for ind,a in enumerate(passover): 
        if passover[ind]>1./3.:
            passout[ind]=1./2.
        else:
            passout[ind]= -1./2.
    return passout

num_flips=1. #Change to how many time the coin is flip.
b=np.arange(1.,num_flips+1.,1.)

num_flips=100.
b=np.arange(1.,num_flips+1.,1.)
coinflip=HeadsOrTails(num_flips,b)

fig=plt.figure(figsize=(8,4))
ax=fig.add_subplot(1,1,1) 

ax.hist(coinflip,label='100 coin flips',bins=2,color='GreenYellow')
ax.set_xlim([-0.5,0.5])
ax.set_title('Probability distribution for unfair coin flips')
ax.set_xlabel('x (Tails if $x=-1/2$, or heads if $x=+1/2$)')
ax.set_ylabel('Occurences')
ax.legend(loc=0,fontsize=10)
```

**4** Perform $N_{flip}$ and take the sample mean $\overline{x}$ of the values.  Do this 2000 times, and show that as $N_{flip}\to\infty$ the distribution of $\overline{x}$ approaches a Normal distribution by plotting the histograms of the means for each.  For $N_{roll}$ choose 2, 16, 64, 256. 

Comment on how the convergence of $\overline{x}$ for the unfair coin compares to the convergence of the sample mean of the dice roll.  
<hr>

```python
def ListSampleMean(num_flips,trials): 
    """takes number of flips ('int') and number trials ('int'). Returns array the size of trials with means for number filps for each trial"""
    passout=np.arange(1.,trials+1.,1.)
    for ind,a in enumerate(passout): 
        b=np.arange(1.,num_flips+1.,1.)
        coinflip=HeadsOrTails(num_flips,b)
        passout[ind]=(1./num_flips)*np.sum(coinflip)
    return passout

num_trials=2000.

ar_num_flips=np.array([2.,16.,64.,256.])
x2=ListSampleMean(ar_num_flips[0],num_trials)
x16=ListSampleMean(ar_num_flips[1],num_trials)
x64=ListSampleMean(ar_num_flips[2],num_trials)
x256=ListSampleMean(ar_num_flips[3],num_trials)

xflips=np.array([x2,x16,x64,x256])

fig=plt.figure(figsize=(8,16))

for nn,me in enumerate(xflips):
    print 'The mean for finding the mean of 2000 \ntrials consisting of ', ar_num_flips[nn], ' coin flips is: ', np.mean(me)
    
    ax=fig.add_subplot(4,1,nn+1) 
    ax.hist(me,bins=80,range=(-0.5,0.5),label='2000 trials',color='DodgerBlue',normed=True)
    ax.set_xlim([-0.5,0.5])
    ax.set_title('Probability distribution of the\nmean for %s  unfair coin flips'%(ar_num_flips[nn]))
    ax.set_xlabel('x (Tails if $x=-1/2$, or heads if $x=+1/2$)')
    ax.set_ylabel('Occurences')
    ax.legend(loc=0,fontsize=12)
fig.tight_layout()
```

After $N_{\text{flips}} \rightarrow 2000$ (taking this to be $N_{\text{flips}} \rightarrow \infty$) the distribution approaches a normal distribution. Where the $\bar{x}_{ N_{\text{flips}}} \rightarrow \mu_x$ for $N_{\text{flips}} \rightarrow \infty$. 

The convergence of $\bar{x}$ for unfair coin flip happens quicker then that of the dice. 

$$\text{% Error for the coin for 2000 flips}= \frac{\mu_{x}-\bar{x}_{2000 \text{ flips}}}{\mu_{x}}=22\%$$

$$\text{% Error for the dice for 2000 flips}= \frac{\mu_{x}-\bar{x}_{2000 \text{ dice}}}{\mu_{x}}=60\%$$
