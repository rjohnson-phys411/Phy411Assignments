### **Q1** Sampling theorem and aliasing

<div style='background:#F0F0F0'>**1** Create a band-limited white noise timeseries with $N=1000$ entries, $dt=0.1 s$, but with all frequencies greater than 0.5 Hz filtered.  Plot the power spectrum to make sure you got it correctly...
</div>

```python
N=1000
dt = 0.1
x0 = np.random.randn(N)
# add a 60 Hz signal to x
t = np.arange(0,N*dt,dt)
#x0 = x0+1.*np.cos(t*2*np.pi*31.5)
```

```python
#Bandwidth = 0.5 Hz
B=0.5
# filter to make Bandlimited signal
fp=B*0.7
fs = B*0.9
fNyq=1/2./dt
n,fn=signal.ellipord(fp/fNyq, fs/fNyq,.1, 60.)
b,a=signal.ellip(n,.1,60.,fn)
xBL = signal.filtfilt(b,a,x0)

fig,ax=plt.subplots(2,1,figsize=(15,6))
plt.axes(ax[0])
ax[0].plot(t,x0,color='0.5',label='full-spectrum signal')
ax[0].plot(t,xBL,'.',color='k',label='band-limited signal')
ax[0].set_xlim([0,30]);ax[0].set_xlabel('time [s]');ax[0].set_ylabel('x [V]');ax[0].legend(fontsize='small',loc=0)
ax[0].legend(fontsize='large')
nfft=10240
pxx,f=mlab.psd(x0,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=mlab.window_hanning)
pxBL,f=mlab.psd(xBL,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=mlab.window_hanning)
plt.axes(ax[1])
ax[1].loglog(f,pxx,color='0.5')
ax[1].loglog(f,pxBL,'.',color='k')
plt.axvline(x=B,linestyle='--',color='k');ax[1].text(B,1,' B=0.5 Hz');ax[1].set_ylim([1e-8,1e2]);ax[1].set_xlabel('f [Hz]');ax[1].set_ylabel('$G_{xx} [V^2 Hz^{-1}]$')
plt.tight_layout()

ax[1].grid(which='major', axis='y')
ax[1].grid(which='both', axis='x')
```

<div style='background:#F0F0F0'>**2** Now decimate by a factor of 4 to get a time series sampled at 2.5 Hz.  Show that you can recover the 10 Hz time series using the Sampling Theorem.</div>

```python
f=2.5
T4=int(1/(dt*f))

y = xBL[range(0,N,T4)]
ty = t[range(0,N,T4)]
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,2))
ax.plot(t,xBL)
ax.plot(ty,y,'.',color='DarkBlue')
ax.set_xlim((60,100))
#ax.set_ylim((-1,1.))
ax.set_ylabel('x(t)[V]')
ax.set_xlabel('t [s]')
ax.legend(('x',' Sampled at %f Hz'%f))
```

COMMENT: 
"*Sampling Theorem states that if $x(t)$ is a bandlimited signal such that $\left|X(f)\right|=0$ for $\left| f\right|>B$, then $x(t)$ can be represented in full by a discrete time series $x_n$ sampled every $\Delta t$ if $\frac{1}{\Delta t} \geq 2B$, and in particular we can reconstruct $x(t)$ as:
$$x(t) = \sum_{n=-\infty}^{\infty} x(n\Delta t) \frac{\sin \left(\pi \left( \frac{t}{\Delta t}- n\right) \right)}{\pi\left(\frac{t}{\Delta t}-n\right)}"$$*
The timeseries has it's frequencies higher then 0.5Hz filter out, so $2B=2(0.5Hz)=1Hz$. Here the data has been sampled at $\frac{1}{\Delta t}= f = 0.25Hz$. Therefore because 2.5Hz > 1Hz the sampling theorem  is valid here. Which can be seen because the dots do a good job at representing the time series.

## Q2: Cubic Spline

Using the data below...

```python
t = np.array([0.,2.5,2.7,3.4,4.2,6.5,9.0])
x = np.cos(t*2*np.pi/3)
```

Perform a cubic spline interpolation onto an even grid from 0 to 9.0 by 
   1. creating the matrix from the notes
   2. inverting and solving for the cubic co-efficients
   3. using those polynomials to evaluate the interpolated data.
Compare with using `scipy.interpolate.interp1d`.

From Notes:

Make the matrix:
\begin{equation} M=
\begin{pmatrix}
    1 & 0 & 0 & \cdots & \cdots & \cdots & \cdots\\
    \Delta_0 & 2\left(\Delta_0+\Delta_1\right) & \Delta_1 & \cdots & \cdots & \cdots & \cdots\\
    \cdots & \cdots & \ddots & \cdots & \cdots & \cdots & \cdots\\
    \cdots & \cdots & \cdots  & \Delta_{N-3} & 2\left(\Delta_{N-3}+\Delta_{N-2} \right) & \Delta_{N-2} & 0\\
    \cdots & \cdots & \cdots &\cdots & \Delta_{N-2} & 2\left(\Delta_{N-2}+\Delta_{N-1} \right) & \Delta_{N-1}\\
    \cdots & \cdots & \cdots & \cdots & \cdots & 0 & 1
\end{pmatrix}
\end{equation}

Where
* $\Delta_i = t_{i+1}-t_i$

```python
dt=np.diff(t) 
dx=np.diff(x)
```

```python
N=len(x)
M=np.zeros((N,N))
#Making matrix M
for n in range(N):
    if n == 0: 
        M[n,n]=1.
    elif 0 < n < N-1:
        M[n,n-1] = dt[n-1]
        M[n,n]   = 2*(dt[n-1]+dt[n])
        M[n,n+1] = dt[n]
    else:
        M[N-1,N-1]=1.
print 'Matrix M is:', M
#invert M
invM=np.linalg.inv(M)
```

Making vector: 
\begin{equation}
\overrightarrow{RHS} = 3\begin{pmatrix}
    0\\
    \frac{dx_1}{\Delta_1}-    \frac{dx_0}{\Delta_0}\\
    \frac{dx_2}{\Delta_2}-    \frac{dx_1}{\Delta_1}\\
    \vdots\\
    \frac{dx_{N-1}}{\Delta_{N-1}}-    \frac{dx_{N-2}}{\Delta_{N-2}}\\   
    0
\end{pmatrix}
\end{equation}
and 
Determining vector $\vec{c}$:
\begin{equation}
\begin{pmatrix}
    c_0\\
    c_1\\
    c_2\\
    \vdots\\
    c_{N-2}\\
    c_{N-1}\\
    c_{N}
\end{pmatrix}
\end{equation}
by $\vec{c}=M^{-1}\overrightarrow{RHS}$

```python
#making RHS vector:
v=np.zeros(N)
for n in range(N):
    if 0 < n < N-1:
        v[n] = 3.*((dx[n]/dt[n])-(dx[n-1]/dt[n-1]))
print 'vector c is:', v
c=invM.dot(v)
```

Need to calculate $q_i(t)=a_i + b_i(t-t_i)+c_i(t-t_i)^2+d_i(t-t_i)^3$. $c_i$ is was just calculated, using $c_i$ the following $a_i, b_i$, and $d_i$ can also be calculate:

\begin{align}
    a_i &=x_i \ \ \ \ \ \ \text{for}\ i=0...N-1\\
    x_i +\Delta_ib_i +\Delta_i^2c_i + \Delta_i^3d_i&=x_{i+1} \ \ \ \ \ \ \text{for}\ i=0...N-1\\
    2c_i+6d_i\Delta_i&=2c_{i+1}\ \ \ \ \text{for}\ i=0...N-2\\
\end{align}

```python
dc = np.diff(c)

a = c[:-1]
d = dc/3/dt
b = dx/dc-c[:-1]*t[:-1]-d*dt**2

q = a+b*dt+c[:-1]*(dt**2)+d*(dt**3)
```

Need to calculate scipy's cubic spline interpolation of x and then plot to compare 

```python
from scipy.interpolate import interp1d
ty = t
t2 = t
x2 = x

ff=interp1d(t2,x2,kind='cubic')
tyn=ty[ty>t2[0]]

tyn=tyn[tyn<t2[-1]]
yi=ff(tyn)


#plot
fig,ax=plt.subplots(1,1,figsize=(10,4))
ax.plot(t,x,'b', label='original')
ax.plot(t,x,'o',color='b')
ax.plot(t[:-1],q,color='g',label='manual cubic spline')
ax.plot(t[:-1],q,'o',color='g')
ax.plot(tyn,yi,color='r',label='scipy interpolate')
ax.plot(tyn,yi,'o',color='r')
ax.set_title('Cubic spline interpolation')
ax.set_ylim((-10,15.))
ax.set_ylabel('x(t)')
ax.set_xlabel('t [s]')
ax.legend()
```

COMMENT: Manual calculated cubic spine isn't a great match. It's ok for a few of the data points. However doesn't fit well for the end points. The scipy.interpolate.interp1d matches the original data well, but has N-2 less data points where end points have been lost.
