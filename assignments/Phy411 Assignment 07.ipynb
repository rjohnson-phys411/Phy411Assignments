```python
N=1001
dt=1.
t=np.arange(0,N*dt,dt)

x = np.cumsum(np.random.randn(N)) # red noise
f0=0.03 # Hz
x = x+ 40.*np.sin(f0*2*np.pi*t) # + sine wave
x=x-np.mean(x)
noise = 60.*np.random.randn(N)
xn = x+noise
xn[N/2:]=xn[N/2:]-2000. #Split time series 
# remove the mean!!
xn = xn-np.mean(xn)
```

<div style='background:#F0F0F0'>**2** Now repeat the above but apply a Hanning window to the frequency window *before* you apply it to the FFT of $x_n$.  Comment on how well the filter is doing when you do this.  (Its helpful to keep the previous result up)
</div>

Q1.2.1

```python
nfft=100 
args = {'NFFT':nfft,'Fs':1./dt,'noverlap':nfft/2.,
        'window':mlab.window_hanning}

#Cal spec of xn
pxn,f=mlab.psd(xn,**args)

#apply filter
fc=0.06 #Hz
X = np.fft.fft(xn)

N = np.shape(xn)[0]
ff = np.linspace(0,1./dt,N)

#make Hanning function. Separate left and right side
han1 = np.hanning(2*np.argmax(ff>fc))
han1R = han1[:len(han1)/2:1]
han1L = han1[len(han1)/2:len(han1):1]

#Swap left and right side. Combine back together to make Hanning final filter.
Zeros = np.zeros(N-len(han1L)-len(han1R))
HanFilter = np.concatenate((han1L,Zeros,han1R))

#reject high frequencies by multiplying freq window by this Hanning filter. 
FilteredX = HanFilter*X

#filtered signal and its spec
y = np.real(np.fft.ifft(FilteredX))+np.mean(xn)
py,fff = mlab.psd(y,**args)

#Plots 
#Plots
fig,ax=plt.subplots(2,1,figsize=(15,6))
plt.axes(ax[0])
ax[0].plot(t,xn,'k',linewidth=1.2,label="signal")
ax[0].plot(t,y,'r',linewidth=1.2,label="filtered")
ax[0].set_xlabel('t [s]')
ax[0].set_ylabel('x [V]')
ax[0].set_title('Signal and the smoothed signal')
ax[0].grid(which='both', axis='y')
ax[0].grid(which='both', axis='x')
ax[0].legend(fontsize='large')

plt.axes(ax[1])
ax[1].plot(t,xn,'k',linewidth=1.2,label="signal")
ax[1].plot(t,y,'r',linewidth=1.2,label="filtered")
ax[1].set_xlim((450,550))
ax[1].set_xlabel('t [s]')
ax[1].set_ylabel('x [V]')
ax[1].set_title('Zoomed signal and the smoothed signal')
ax[1].grid(which='both', axis='y')
ax[1].grid(which='both', axis='x')
ax[1].legend(fontsize='large')


# spectrum
#plt.axes(ax[2])
#ax[2].loglog(f,pxn,color="0.4",label="signal")
#ax[2].loglog(f,py,'r',linewidth=1.2,label="filtered")
#ax[2].set_xlabel('f [Hz]')
#ax[2].set_ylabel(r'$G_{xx} \mathrm{[V^2 Hz^{-1}]}$')
#ax[2].set_title('spectrum')
#ax[2].legend(loc=3,fontsize='large')
plt.tight_layout()
```

```python
# Time Series:
N=20001
dt=1.
t=np.arange(0,N*dt,dt)
x = np.cumsum(np.random.randn(N)) # red noise
x=x-np.mean(x)
```

```python
# Calculating the truncated Dirichlet Kernel to filter at 0.15 Hz low-pass:
k = np.array(range(-N/2+1,N/2))
fcL = 0.15
M = fcL*2*N+1
hL2 = (1./N)*np.sin(np.pi*(k+0.000001)*M/N)/np.sin(np.pi*(k+0.000001)/N)

filtlen = 40
hL = hL2*1.
hL = hL[abs(k)<filtlen+1]

# Calculating the truncated Dirichlet Kernel to filter at 0.095 Hz low-pass to make high-pass
fcH = 0.095
M = fcH*2*N+1
h = (1./N)*np.sin(np.pi*(k+0.000001)*M/N)/np.sin(np.pi*(k+0.000001)/N)

#The following code didn't seem to make a high-pass filter:
#hH2 = hL2*0.
#for ind, num in enumerate(hL2):
#    hH2[ind]=(-1)**ind*num

#found this code online that seems to work 
#high-pass filter from the low-pass filter through spectral inversion.
hH2 = -h
hH2[N/2] += 1

hH = hH2*1.
# how long is the filter:
hH = hH[abs(k)<filtlen+1]
kFilt=k[abs(k)<filtlen+1]

#If I add the following code to try fix the discontinuity in the
#high-pass filter the filter doesn't seem to work, so I've commented it out:

#hL[len(hL)/2]=M/N
#hL2[len(hL2)/2]=M/N


fig,ax=plt.subplots(1,1,figsize=(15,4))
ax.plot(k,hH2,'DarkRed',linewidth=0.5)
ax.plot(kFilt,hH,'r',linewidth=2,label='%f Hz low pass'%fcL)
ax.plot(k,hL2,'DarkBlue',linewidth=0.5)
ax.plot(kFilt,hL,'b',linewidth=2,label='%f Hz high pass'%fcH)
ax.set_xlabel('k')
ax.set_ylabel('$h_k$')
ax.set_xlim([-50,100])
ax.set_title('Filters')
ax.legend()
```

```python
#Combining, or convolving, the low- and high-pass filters to make a band-pass filter.
#This convolves a low-pass filter with fc of 0.15Hz and a high-pass filter with fc of 0.095.
#This makes band-pass filter centered at 0.1Hz with 0.1 Hz wide.
filt = np.convolve(hL,hH)
y = np.convolve(x,filt)
```
