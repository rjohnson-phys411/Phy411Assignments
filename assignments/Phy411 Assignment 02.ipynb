## Q1: Show that the correlation co-efficient is indeed related to the fraction of varaince explained by a linear dependence.

**1** To start, make a $N(0,1)$ random variable $x$, with 200,000 entries.  Then make a linearly dependent variable $y_i= a x_i + n_i$, where a is some amplitude, and $n_i$ is a random noise varaible.  Assume that $n_i$ is distributed as $N(0,n)$, where $n$ is the noise level.

Calculate the correlation co-efficient $r_{xy}$ between $x$ and $y$ for 40 different values of $n$, logarithmically spaced between 0.01 and 100. (hint use `np.logspace`).  

Plot $r_{xy}$ versus the inverse of the noise $n$ normalized by the standard devaition of the signal $ax_i$: $\frac{a s_x}{n}$   You may want to use `ax.loglog` to visualize the orders of magnitude.  

Comment on the resulting comparison.
<hr>

Approximate correlation co-efficient:
\begin{equation}
      r_{xy} = \frac{s_{xy}}{s_xs_y} =\frac{1}{N-1}\frac{\sum_{i=1}^N (x_i-\overline{x})(y_i-\overline{y})}{s_xs_y}
\end{equation}

```python
N= 200000;m= 40;a=1.

#making logarithmically spaced values of ntosig
ntosig= np.logspace(-2,2,num=m)
x= np.random.randn(N)

#make r array to copy values to
r=np.zeros(m)
#calculating r for the 40 values of n  
for ind,c in enumerate(ntosig):
    y = a*x + np.random.randn(N)*c
    r[ind] = np.sum(x*y)/np.std(x,ddof=1)/np.std(y,ddof=1)/(N-1)
    
#inverse of the noise n normalized by the standard deviation of the signal
inofnoise=a*np.sqrt(np.std(x,ddof=1)/(m-1))/ntosig
```

```python
#plot
fig=plt.figure(figsize=(15,5))
ax=fig.add_subplot(1,1,1) 
ax.loglog(inofnoise,r,label='',color='b')
ax.set_ylim(0.005,1.1)
ax.set_xlabel(r'$\frac{as_x}{n}$',fontsize=25)
ax.set_ylabel('$r_{xy}$',fontsize=25)
ax.set_title('Approximate correlation co-efficient vs. Approximate correlation co-efficient',fontsize=20)
ax.tick_params(axis='both', labelsize=18)
```

The less noise there is present in the data (i.e. larger the inverse of the noise is) then the stronger the covariance is between the two variables $x_i$ and $y_i$. In this case a stronger linear association between $x_i$ and $y_i$ (i.e. linear correlation co-efficient,  $r_{xy} \rightarrow 1$) when $n \rightarrow 0$.

**2** Repeat the above, but with a different value of $a$.  Comment on how (and why) $a$ changes the value of $r_{xy}$.  

```python
#same as previous, but with a=10
N= 200000;m= 40;a=10.

ntosig= np.logspace(-2,2,num=m)
x= np.random.randn(N)

r=np.zeros(m)

for ind,c in enumerate(ntosig):
    y = a*x + np.random.randn(N)*c
    r[ind] = np.sum(x*y)/np.std(x,ddof=1)/np.std(y,ddof=1)/(N-1)
    
inofnoise=a*np.sqrt(np.std(x,ddof=1)/(m-1))/ntosig
```

```python
#plot
fig=plt.figure(figsize=(15,5))
ax=fig.add_subplot(1,1,1) 
ax.loglog(inofnoise,r,label='',color='b')
ax.set_ylim(0.09,1.1)
ax.set_xlabel(r'$\frac{as_x}{n}$ $\left[\frac{1}{V}\right]$',fontsize=20)
ax.set_ylabel('$r_{xy}$',fontsize=25)
ax.set_title('Approximate correlation co-efficient vs. Approximate correlation co-efficient',fontsize=20)
ax.tick_params(axis='both', labelsize=18)
```

Here $x_i$ in $y_i$ has started off by being multiplied by $a=10$, which makes the noise term $n_i$ less significant then $x_i$ for $y_i$. Thus linear correlation co-efficient,  $r_{xy}$, reaches 1 at a quicker rate then the previous graph ($a=1$). This because $x_i$ and $y_i$ started off with a greater covariance, in this case a stronger linear association, then that of $y_i$ and $n_i$, thus starting at value of $r_{xy}$ that is closer to 1 then the previous plot.

## Q2: Monte Carlo on slopes and test versus student-t distribution

**1** We stated without proof that the slope $b$ of a linear fit of $y$ to $x$ is distributed as a student-t distribution with $N-2$ degrees of freedom, with a standard devaition given by the standard deviation of $y$ around the fit model $\tilde{y}$.   Lets test that emperically using a Monte Carlo technique.  

First create an independet variable $x$ from 10 points from an $N(0,1)$ distribution.

Next, using $n=1.0$, and $a=1.0$ create the dependent variables $y_i=a x_i + n_i$ where $n_i$ are drawn from the $N(0,n)$ distribution.  Calculate the slope $b$.  

Do this 10000 times, and make a histogram of the resulting fits $b$.  Also histogram $s_{y|x}$ as defined in the notes.

Comment on the histigrams

<hr>

The $s_{y|x}$ sample standard deviation is:
\begin{equation}
     s_{y|x} = \left[\frac{\sum_{i=1}^N \left( y_i-\tilde{y}_i \right)^2 }{N-2} \right]^{1/2}
\end{equation}

```python
N=10;times=10000;ntosig=1.0

b= np.zeros(times)
syx= np.zeros(times)
#calculating slope b and sample standard deviation syx 10000 times 
for ind in range(times):
    a=1.0
    x= np.random.randn(N)
    y = a*x + np.random.randn(N)*ntosig

    bt = np.sum(x*y)-N*np.mean(x)*np.mean(y) #bt= b temp
    bt = bt/(np.sum(x**2)-N*np.mean(x)**2)
    b[ind]= bt
    
    a = np.mean(y)-bt*np.mean(x)
    yhat=bt*x+a
    syx[ind]=np.sqrt(np.sum((y-yhat)**2)/(N-2))

    ind= ind+1
```

The fits slope, or *fits* $b$ is:
\begin{align*} 
    \text{fits }b &=\left(b-E(b)\right) \left( \sum_{i=1}^N \left(x_i-\overline{x}\right)^2\right)^{1/2}
\end{align*} 

```python
#Calculating fits b. Some reason python didn't like me doing it as one calculation.
#When i broke it up in step seems to work. Probably just me. 
Eb=np.mean(b)
summ=x-np.mean(x)
summ=summ**2
summ=np.sum(summ)
summ=np.sqrt(summ)
#anyway, now we have fits b
bfit=(b-Eb)*(summ)
```

```python
#plot
fig=plt.figure(figsize=(15,5))
ax=fig.add_subplot(1,1,1)
ax.hist(bfit,label='fit $b$ [no units]',bins=100,normed=True,histtype='step',color='b')
ax.hist(syx,label='$s_{y|x}$',bins=100,normed=True,histtype='step',color='g')
ax.set_xlabel('measurement');ax.set_ylabel('Occurences');ax.set_title('')
ax.set_title('histograms for fit $b$ and $s_{y|x}$' )
ax.legend(loc=0,fontsize=12)
```

Here we can see that histogram for *fit* $b$ looks like a student-t probability density function. Where histogram looks more like a maybe a normal distribution, with $\mu=1$.

"The slope b of a linear fit of y to x is distributed as a student-t distribution with N−2 degrees of freedom", meaning:
\begin{align*} 
    \text{fits }b =\left(b-E(b)\right) \left( \sum_{i=1}^N \left(x_i-\overline{x}\right)^2\right)^{1/2} &= s_{y|x} t_{N-2},  \;\text{ expected value for $b$ is } E(b)= \bar{x} \\

\end{align*} 

```python
n=N-2 #degrees of freedom
x = np.linspace(-10,10.,1000)

#plot
fig=plt.figure(figsize=(15,5))
ax=fig.add_subplot(1,1,1)
y=stats.t.pdf(x,n)
ax.plot(x,y,label='$t_{%2d}$ distribution'%n,color='r')
ax.hist(bfit/syx,label=r'$\frac{Fit\, b}{s_{y|x}}$',bins=200,normed=True,histtype='step',color='b',alpha=0.8)
ax.set_title('Comparison of $t_{%2d}$ distribution model and histogram of the resulting scaled fits $b$'%n, fontsize=18)
ax.set_xlabel(r'Measurement of $\frac{Fit\, b}{s_{y|x}}$',fontsize=20)
ax.set_ylabel('Occurrences', fontsize=18)
ax.legend(fontsize=18)
```

## Q3: Linear comparison of Deep Cove data to James Bay temperature data

```python
# Load the data:

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc= hourdata[0,:]
jb= hourdata[1,:]
```

**1** Make a scatterplot of Plot James Bay versus Deep Cove.

```python
#removing incomplete data points that don't have a matching readings for jb and dc for the same timestamp. 
#i.e. one of temp. readings for a give time for jb or dc is a nan value. 
#Thus both are not used.
dcjb= np.vstack([dc,jb]) #Merging the two arrays
dcjb= dcjb.T 
dcjb= dcjb[~np.isnan(dcjb).any(1)] #removing any rows with nan values
dc_nonan,jb_nonan= np.hsplit(dcjb, 2) #splitting columns into two new arrays
x= dc_nonan.T[0]; y= jb_nonan.T[0] #putting data backing original array format
#note: x=dc and y=jb
```

**2** Compute the linear Correlation Co-efficient.

```python
#need to find the number values in x or y (should be the same)
N= np.shape(x)[0]
#calculating appox. linear correlation co-efficient
y= y-np.mean(y); x= x-np.mean(x)
r= np.sum(x*y)/np.std(x,ddof=1)/np.std(y,ddof=1)/(N-1)
print 'Approximate of linear Correlation Co-efficient ,r , is', r 
```

**3** Fit a straight line to the James Bay data as a fucntion of the Deep Cove data.  Comment on the fit.  (Do **NOT** use a canned routine for the fitting)

```python
#perform a linear regression
b = np.sum(x*y)-N*np.mean(x)*np.mean(y)
b = b/(np.sum(x**2)-N*np.mean(x)**2)
a = np.mean(y)-b*np.mean(x)
yhat=b*x+a

#plotting previous plot now with yhat(x)
fig=plt.figure(figsize=(15,8))
ax=fig.add_subplot(1,1,1)
ax.plot(x,y,'.')
ax.plot(x,yhat,'-',label='Fit: y=%1.4f+%1.4fx'%(a,b), color='r')
plt.legend(loc=0)
ax.set_xlabel(labelx);ax.set_ylabel(labely)
ax.set_title('Temperature of James Bay versus Deep Cove.')
```

**4** Add the confidence intervals to your fit, and comment.  ***NO CANNED ROUTINES***

```python
#plotting previous plot, but now adding the confidence in the slope
fig=plt.figure(figsize=(15,8))
ax=fig.add_subplot(1,1,1)
ax.plot(x,y,'.')
ax.plot(x,yhat,'-',label='Fit: y=%1.4f+%1.4fx'%(a,b),linewidth=0.5, color='r')
plt.legend(loc=0)
ax.set_xlabel(labelx);ax.set_ylabel(labely)
ax.set_title('Temperature of James Bay versus Deep Cove.')

# get the error bounds...
syx=np.sqrt(np.sum((y-yhat)**2)/(N-2))

bdn,bup=stats.t.interval(0.95,df=N-2,scale=syx)
bb=np.array([bdn,bup])
bb=bb/np.sqrt(np.sum((x-np.mean(x))**2))
print 'Slope b is between %f and %f with 95%% probability'%(b+bb[0],b+bb[1])
xx=np.sort(x)
yhathigh=(b+bb[1])*(xx-np.mean(x))+np.mean(y)
yhatlow=(b+bb[0])*(xx-np.mean(x))+np.mean(y)
ax.fill_between(xx,yhathigh,yhatlow,alpha=0.8,color='Lime')
```

Note: The confidence in the slope is quite high; thus hard to see them in this plot.

Here we can see that James Bay's temp. will probability increase between 0.753791 and 0.759110 times with every increase of Deep Cove's temp, with a confidence level of 95%.

**NOTE** There is a way to treat both axes independently called "Neutral Regression".  It is relatively simple, and involves normalizing the total distance of each point from the trend line. <a href="http://journals.ametsoc.org/doi/pdf/10.1175/1520-0426(1999)016%3C0876%3AAPFANR%3E2.0.CO%3B2">(Marsden 1999)</a>. This technique is strongly preferred if there is no reason to think that one data set is dependent on the other.  A caveat is that the data sets shoudl be normalized before fitting so that the "distance" has meaning.
