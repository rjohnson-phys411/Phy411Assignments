### Error bounds on coherence estimates

<div style='background:#F0F0F0'>  The 95% confidence intervals for the coherence squared $\gamma^2_{xy}(f)$ between two variables $x(t)$ and $y(t)$ are approximately:
$$ \gamma^2_{xy} \pm \frac{\sqrt{2}\left(1+\gamma^2_{xy}\right)}{\left|\gamma_{xy} \right|\sqrt{n_d}}$$
where $n_d$ is the number of blocks used to calculate the coherence.  

You **may** use `matplotlib.mlab.csd` for the following question, but you had best use it correctly!
</div>

<div style='background:#F0F0F0'>**1** Make an input process $x(t)$ as random Gaussian noise, and the output process the sum of a noise source and a running average of the input: $y(t)=n(t)+\frac{1}{b}\int_{t-b}^t x(t')dt'$ and plot (Zoom so we can see whats going on, and label your plot).  Make sure you generate a large time series.</div> 

The code to my code. ...left is the terms in the equation in the question, right side of := is the code that repressents the same term. Followed by it's spectrum. 

* x(t) := Force ("input") = F --> pf = G_FF
* RunningAvg := x --> px = G_xx
* n(t) := n --> pn = G_nn
* y(t) := xn ("output") --> pxn = G_xn
* cFxn --> G_Fxn 

```python
N=25600
b=100.
t=np.arange(N)
mu, sigma = 0, 1 # mean and standard deviation
F = np.random.normal(mu, sigma, N) #x(t)
x =(1/b)*np.convolve(F,np.ones((b,)),mode='same') #RunningAvg
n =np.random.rand(len(x))
n = (n - np.mean(n))*0.1 #n(t)
xn = n+x #y(t)

fig,axs=plt.subplots(2,1,figsize=(15,10))


plt.axes(axs[0])
axs[0].plot(t,F,label='Input',color='b',linewidth=1.)
axs[0].plot(t,xn,label='Output',color='DarkBlue',linewidth=1.)
axs[0].set_xlim(0,100)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Signal (V)')
axs[0].legend(loc=1,fontsize='large')

plt.axes(axs[1])
axs[1].plot(t,F,label='Input',color='b',linewidth=0.2)
axs[1].plot(t,xn,label='Output',color='DarkBlue',linewidth=0.2)
axs[1].set_xlim(0,25600)
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Signal (V)')
axs[1].legend(loc=1,fontsize='large')

leg = plt.gca().get_legend()
llines = leg.get_lines()   
plt.setp(llines, linewidth=2.5)
plt.show()
```

<div style='background:#F0F0F0'>**2** Compute the coherence squared of the input and the output.  Make a plot with the spectra in the upper panel

and the coherence squared in the middle panel

and the phase in the bottom.

You will need to choose a value for your fft length.  Choose a relative low one to get a smooth plot.
</div>

$$ 95\%\text{ confidence interval for } \gamma^2_{xy} = \gamma^2_{xy} \pm \frac{\sqrt{2}\left(1+\gamma^2_{xy}\right)}{\left|\gamma_{xy} \right|\sqrt{n_d}}$$

```python
dt=1.
NFFT = 512
nd=N/NFFT*1.
print 'number of blocks, sqrt(n_d) =', nd


cFxn,f=mlab.csd(xn,F,NFFT,Fs=1./dt,window=mlab.window_none,noverlap=256)
cxxn,f=mlab.csd(xn,x,NFFT,Fs=1./dt,window=mlab.window_none,noverlap=256)

pxn,fxn=mlab.psd(xn,NFFT,Fs=1./dt,window=mlab.window_none,noverlap=256)
pF,fxn=mlab.psd(F,NFFT,Fs=1./dt,window=mlab.window_none,noverlap=256)
pn,fxn=mlab.psd(n,NFFT,Fs=1./dt,window=mlab.window_none,noverlap=256)
px,fxn=mlab.psd(x,NFFT,Fs=1./dt,window=mlab.window_none,noverlap=256)

fig,axs=plt.subplots(3,1,figsize=(15,10))

plt.axes(axs[0])
axs[0].loglog(f,pF,label='Forcing', color='b')
axs[0].loglog(f,px,label='Response w/o Noise', color='g')
axs[0].loglog(f,pn,label='Noise', color='r')
axs[0].loglog(f,pxn,label='Response w/ Noise', color='Cyan')
axs[0].set_ylim((1e-8,100.))
axs[0].set_ylabel(r'$G(f)(V^2Hz)^{-1})$')
axs[0].legend(loc=0,fontsize='small')

plt.axes(axs[1])
CoSq=np.abs(cxxn)**2/px/pxn
CoSq95_LeftTerm=np.sqrt(2.0)*(1.0+CoSq)/np.abs(CoSq)/np.sqrt(nd)
CoSq95min=CoSq-CoSq95_LeftTerm
CoSq95max=CoSq+CoSq95_LeftTerm

axs[1].semilogx(f,CoSq,'k')
axs[1].fill_between(f,CoSq95min,CoSq95max,alpha=0.5,
    linewidth=0.0,edgecolor=None,color=None,facecolor='b')

one=np.ones(len(CoSq))
zero=np.zeros(len(CoSq)) 
axs[1].semilogx(f,one,'r',alpha=0.5)
axs[1].semilogx(f,zero,'r',alpha=0.5)
axs[1].set_ylim((-2.,2.))
#axs[1].set_ylim((0,1.1))
axs[1].set_ylabel(r'$\gamma^2_{xy}$')

plt.axes(axs[2])
ang = np.angle(cxxn)
axs[2].semilogx(f,ang/np.pi,'k')
axs[2].set_yticks(np.arange(-1,1,0.5))
axs[2].set_ylabel('$Phase_{xy} / \pi$')
axs[2].set_xlabel('$f / f_N$')
```

COMMENT:

In the first plot the Forcing, or input, spectra ($G_ff$) is in blue. The output spectrum is in Cyan ($G_nn$). Without the added noise (it's spectra, $G_nn$ is in red) the response spectra from the forcing is $G_xx$ and is in green. When the noise spectra is greater then the response spectra then the noise spectra dominates. This can be seen in the response spectra with noise ($G_yy$), or the output signal spectra. 

The second plot illustrates this effect. If we're just looking at the plot of $\gamma^2_{xy}$, then it starts out at 100% (squared) coherent. However as the response declines for higher frequencies it gets closer to the same intensity as the noise spectra the coherence starts to very and decline. As the noise spectra becomes more and more dominant the coherence squared, and thus the coherence, approaches zero.

However considering coherence squared is measured from 0 to 1 (these limits in red in the plot), the 95% confidence interval in the coherence squared is quite spread out. When the noise becomes the dominant spectra the confidence interval starts to spread out quite drastically. So much that coherence squared between the response and the output for the noise dominant region is virtually unknown. 

For the phase plot the response is in phase with the output signal until the noise spectra starts becoming dominant. After this point the phase starts jumping around when these two signal become incoherent. However since the phase is based off of coherence squared values there might be some variance in the confidence of this phase plot.

</div>

```python
def PlotGraphs(NFFT,numNFFT):
    fig,axs=plt.subplots(3,numNFFT,figsize=(20,10))
    for i in range(numNFFT):
        dt=0.1
        nd=N/NFFT[i]*1.
        print 'number of blocks, sqrt(n_d) =', nd, 'for plots in row', i+1

        
        cFxn,f=mlab.csd(xn,F,NFFT[i],Fs=1./dt,window=mlab.window_none,noverlap=256)
        cxxn,f=mlab.csd(xn,x,NFFT[i],Fs=1./dt,window=mlab.window_none,noverlap=256)

        pxn,fxn=mlab.psd(xn,NFFT[i],Fs=1./dt,window=mlab.window_none,noverlap=256)
        pF,fxn=mlab.psd(F,NFFT[i],Fs=1./dt,window=mlab.window_none,noverlap=256)
        pn,fxn=mlab.psd(n,NFFT[i],Fs=1./dt,window=mlab.window_none,noverlap=256)
        px,fxn=mlab.psd(x,NFFT[i],Fs=1./dt,window=mlab.window_none,noverlap=256)

        plt.axes(axs[0,i])
        axs[0,i].loglog(f,pF,label='Forcing')
        axs[0,i].loglog(f,px,label='Response w/o Noise')
        axs[0,i].loglog(f,pn,label='Noise')
        axs[0,i].loglog(f,pxn,label='Response w/ Noise')
        axs[0,i].set_ylim((1e-8,100.))
        axs[0,i].set_ylabel(r'$G(f)(V^2Hz)^{-1})$')
        axs[0,i].set_title('For NFFT = %d (nblocks=%d)'%(NFFT[i],nd))
        axs[0,i].legend(loc=0,fontsize='small')

        plt.axes(axs[1,i])
        CoSq=np.abs(cxxn)**2/px/pxn
        CoSq95_LeftTerm=np.sqrt(2.0)*(1.0+CoSq)/np.abs(CoSq)/np.sqrt(nd)
        CoSq95min=CoSq-CoSq95_LeftTerm
        CoSq95max=CoSq+CoSq95_LeftTerm


        axs[1,i].semilogx(f,CoSq,'k')
        axs[1,i].fill_between(f,CoSq95min,CoSq95max,alpha=0.5,
            linewidth=0.0,edgecolor=None,color=None,facecolor='b')

        one=np.ones(len(CoSq))
        zero=np.zeros(len(CoSq)) 
        axs[1,i].semilogx(f,one,'r',alpha=0.5)
        axs[1,i].semilogx(f,zero,'r',alpha=0.5)
        axs[1,i].set_ylim((-2.,2.))
        #axs[1,i].set_ylim((0,1.1))
        axs[1,i].set_ylabel(r'$\gamma^2_{xy}$')

        plt.axes(axs[2,i])
        ang = np.angle(cxxn)
        axs[2,i].semilogx(f,ang/np.pi,'k')
        axs[2,i].set_yticks(np.arange(-1,1,0.5))
        axs[2,i].set_ylabel('$Phase / \pi$')
        axs[2,i].set_xlabel('$f / f_N$')
        
NFFTarray = np.array([512,1536,4608])
PlotGraphs(NFFTarray,len(NFFTarray))
```

Looking at the plot, the mean acceleration of the vehicle containing the shear probes was constant. However the mean turbulence shear seem to be zero. As the vehicle accelerated quickly it added to the variance in the data from the turbulence shear probe. However a slower change in the vehicle's acceleration didn't seem to affect the turbulence shear probe.

<div style='background:#F0F0F0'> **2** Compute the power spectra of the two signals and the coherence squared (you could do the phase too, but it is not too thrilling).  Comment on the spectral similarities and the coherence.  </div>

```python
x = shear
F = ac
#---------
dt = np.max(time)/len(time)
NFFT = 512
nd=N/NFFT*1.


cFx,f=mlab.csd(x,F,NFFT,Fs=1./dt,window=mlab.window_none,noverlap=256)
pF,fxn=mlab.psd(F,NFFT,Fs=1./dt,window=mlab.window_none,noverlap=256)
px,fxn=mlab.psd(x,NFFT,Fs=1./dt,window=mlab.window_none,noverlap=256)

fig,axs=plt.subplots(3,1,figsize=(12,9))

plt.axes(axs[0])
axs[0].loglog(f,px,label='Response',color='b')
axs[0].loglog(f,pF,label='Forcing',color='DarkBlue')
axs[0].set_ylim((1e-8,100.))
axs[0].set_ylabel(r'$G(f) (s^2Hz)^{-1}$')
axs[0].legend(loc=0,fontsize='small')

plt.axes(axs[1])
CoSq=np.abs(cFx)**2/pF/px
axs[1].semilogx(f,CoSq,'k')


axs[1].set_ylim((-5,5))
#axs[1].set_ylim((,1.1))
axs[1].set_ylabel(r'$\gamma^2$')

plt.axes(axs[2])
ang = np.angle(cFx)
axs[2].semilogx(f,ang/np.pi,'k')
axs[2].set_yticks(np.arange(-1,1,0.5))
axs[2].set_ylabel('$Phase / \pi$')
axs[2].set_xlabel('$f / f_N$')
```

COMMENT:
Here the 'force' is the acceleration in the vehicle. The response is the measured turbulence shear by the probe. For the lower frequencies there is no coherence (or coherence squared) between the acceleration of the vehicle and the turbulence shear measured by the probe. However at the higher end of the spectra there is some coherence. In particular the spikes in the response, mainly the two spikes at around $f=10.5Hz$ and $14Hz$. Hinting that these are resonance frequencies for the acceleration of the vehicle and the probe readings at these frequencies.
