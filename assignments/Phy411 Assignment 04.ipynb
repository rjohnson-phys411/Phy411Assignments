**2** Write a second function to return the estimate of the power spectrum of $x$.  For posterity sake, include a value of the sampling frequency, and return the proper frequencies for the estimate.  Plot the resulting spectrum.

\begin{equation}
    \frac{n\hat{G}_{xx}}{\left. \chi^2_n \right|_{0.975}} \leq {G_{xx}} \leq \frac{n\hat{G}_{xx}}{\left. \chi^2_n \right|_{0.025}}
\end{equation}
$n=$ degrees of freedom $=2$

````python
def eps(G):
    """Returns chi squared evaluated at 0.975 (highest) and 0.025 (lowest)  for n degrees of freedom = 2"""
    import scipy.stats as stats
    inter = stats.chi2.interval(0.95,df=2)
    highEst = 2.*G[0:]/inter[1]
    lowEst = 2.*G[0:]/inter[0]
    return highEst,lowEst
````

````python
#Calculating power spectrum of x (Gxx) and it's spectral estimate from chi^2_2
T = 2048. 
dt = T/N

#Calculating Gxx
f = np.arange(N/2)/T
X = np.fft.fft(x)
Gxx = np.real(dt*np.conj(X[:N/2])*X[:N/2])

#spectral estimate
highEst,lowEst = eps(Gxx)
````

````python
#plot
fig=plt.figure(figsize=(12,4))
ax=fig.add_subplot(1,1,1) 
ax.loglog(f[0:],Gxx[0:], label=r'Gxx (with 95% confidence bounds from $\chi^2_2$',color='b')
ax.fill_between(f[0:],lowEst,highEst,alpha=0.5,
                linewidth=0.0,edgecolor=None,color=None,facecolor='b')
ax.set_xlabel(r'$f \ [Hz]$')
ax.set_ylabel(r'$G_{xx}(f_k) \ [(V)^2Hz^{-1}]$')
ax.set_title('Estimate of the power spectrum for red noise')
ax.legend()
````

**3** Now do the Monte-Carlo, by running the above 100 times and plotting the resulting spectral estimates as dots on the same plot. As you are making your data, save it into an array.  I did something like:

````python
#'n=numbers over runs':
n = 100
Gxs = np.zeros((100,500))
#calculating Monte-Carlo for spectral estimates. stored in Gxs.
for ind in range(n):
    N= 1000 #
    x = np.cumsum(np.random.randn(N))
    t = np.arange(N)

    T = 2048.
    dt = T/N

    f = np.arange(N/2)/T
    X = np.fft.fft(x)
    G = np.real(dt*np.conj(X[:N/2])*X[:N/2])
    Gxs[ind,:]=G
````

````python
#plot
fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(1,1,1) 
for ind in range(n):
    ax.loglog(f[0:],Gxs[ind],'.',markersize=2) 
ax.set_xlabel(r'$f \ [Hz]$')
ax.set_ylabel(r'$G_{xx}(T,f) \ [(V)^2Hz^{-1}]$')
ax.set_title('Monte-Carlo of the power spectrum for red noise (100 trials)')
````

````python
GxsMean = np.mean(Gxs, axis=0)

GxsSort = np.sort(Gxs, axis=0)
GxsLow = np.zeros(500)
GxsHigh = np.zeros(500)

for ind in range(500):
    GxsHigh[ind] = GxsSort[3-1,ind] #because 0.025N=0.025(100)=2.5 it has to be rounded because there's not [2.5,ind] element in GxsSort. I'll round up to 3.
    GxsLow[ind] = GxsSort[98-1,ind] #same goes for 0.975*100=97.5. I'll round up to 98.
````
