**1** From the hourly time series at Deep Cove (`dc`), calculate the *lag correlation co-efficient* $\rho_{xx}(\tau)$ out to a lag of $\tau=30\ \mathrm{days}$ (Umm, remember this data is hourly!).  (also rememebr to remove the mean of $dc$ before trying to to the calculations).  

Comment on the resulting plot.

````python
#drop nan values
dc = dc[~np.isnan(dc)]
````

````python
#lag of \pm 30 days:
lags = np.arange(-720,720.)
#make x un-bias:
x = dc-np.mean(dc)

#cxx (for un-bias x) from the notes:
cxx = 0.*lags
for ind,tau in enumerate(lags):
    if tau<0:
        cxx[ind] = np.mean(x[:tau]*x[-tau:])
    elif tau==0:
        cxx[ind] = np.mean(x*x)
    else:
        cxx[ind] = np.mean(x[:-tau]*x[tau:])
        
#Plot:
fig=plt.figure(figsize=(15,6))
ax=fig.add_subplot(1,1,1) 
ax.plot(lags,cxx/np.var(x),label='30 day lag correlation')
ax.legend()
ax.set_xlabel(r'$\tau [h]$')
ax.set_ylabel(r'$\rho_{xx}$')
ax.set_title('Lag correlation co-efficient for Deep Cove')
````

The small oscillation in $\rho_{xx}(\tau)$ is correlation of temp at $\tau=0$ to that of the daily cycle of temperature. When 24h passes the oscillation is at max because the temperature has comeback to the closest temperature of when $\tau=0$. 

The general trend of $\rho_{xx}(\tau)$ is quite different then the 24h cycle. The lag correlation drops off quite quickly for about 5 days. Then the drop in the general correlation slows down to almost a linear fashion, which is still decreasing.

**1** Consider the process $x(t) = \int_{t-b}^t y(t) \ \mathrm{dt}$ where $y(t)$ is uncorrelated white noise so that $R_{yy}(\tau)=\delta(\tau)$.  Compute $R_{xx}(\tau)$.

So we have $x(t) = \int_{t-b}^t y(t) \ \mathrm{dt}$. Where $y(t)$ is uncorrelated white noise such that $R_{yy}(\tau)=\delta(\tau)$

From the notes $R_{xx}(\tau)$ is:
\begin{align}
    R_{xx}(\tau) &= E\left[ x(t)x(t+\tau) \right] \\
        &= \int_{-\infty}^{-\infty} x(t)x(t+\tau)  \ \mathrm{dx} \\
        &= \int_{-\infty}^{-\infty} p(x) \left( \int_{t-b}^{t} y(t_1) \ \mathrm{dt_1} \int_{t+\tau-b}^{t+\tau} y(t_2) \ \mathrm{dt_2} \right) \ \mathrm{dx} \\
    R_{xx}(\tau) &= \int_{t-b}^{b} \int_{t+\tau-b}^{t+\tau} E\left[ y(t_1)  y(t_2) \right] \ \mathrm{dt_1} \ \mathrm{dt_2}
\end{align} 

However,
$$E\left[ y(t_1)  y(t_2) \right]=R_{y_1y_2}=\delta(t_2-t_1)=\left\{
  \begin{array}{l l}
    1 & \quad \text{if } t_1 = t_2\\
    0 & \quad \text{else}
  \end{array} \right.$$

Here we have two boxcar functions of time. Both of length -b and separated by $\tau$, thus:

$$R_{xx}(\tau) = \left\{
 \begin{array}{l l 1 1}
    b\mp\tau & \quad \text{if }0<\tau<\pm b\\
    b & \quad \text{if }\tau = 0\\
    0 & \quad \text{else}
  \end{array} \right.$$

When $\tau=0$ there is a complete overlap of the boxcar functions and therefore $R_{xx}=1$. However when $|\tau|> b$ there is no overlap and $R_{xx}=0$. When in between this two states there's two related linear relationship as shown above.

````python
def MakeFuncX(b,N):
    """Makes function 'x(t)' defined in the question. Input: 'b (int)' and 'N (int)'. Output: function 'x (array)' of size 'N'."""
    #makes box function of length b:
    bArray = np.ones(b)
    #makes function of white noise:
    y = np.random.randn(N)
    #convolve bArray and y to do integratal x:
    x = np.convolve(y,bArray,mode='same')
    return x

def sdf(x,f,N):
    """Calculates the finite spectral density function. Input: 'x (array'), 'f (array)', and 'N (int). Output: 'SxxT(array)'"""
    t = np.arange(0.,N)
    fsize = np.shape(f)[0]
    SxxT = np.zeros(fsize)
    for ind, F in enumerate(f):
        #Finite fourier transform:
        X = sum(x*np.exp(-2.j*np.pi*F*t))
        conX = X.conjugate()
        SxxT[ind] = X*conX/N
    return SxxT

def CalSxxT(b,f,N):
    """Calculates all SxxT for list of N values. Stacks these arrays of SxxT into SxxTArray. Input: 'b (int)', 'f (array)', 'N (array)'. Output: 'SxxTArray(array)'"""
    for i in range(np.shape(N)[0]):
        x = MakeFuncX(b,N[i])
        if i == 0:
            SxxTArray = sdf(x,f,N[i])
        else:
            SxxTArray = np.vstack((SxxTArray,sdf(x,f,N[i])))
    return SxxTArray
````

````python
N = [500., 5000., 50000.]#, 5e6]
b = 100
f = np.logspace(-4,-0.1,5000)
        
#makes array of which the rows are just SxxT for each value of N
SxxTArray = CalSxxT(b,f,N)
````

````python
#Plot SxxT for each N
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(1,1,1) 
for i in range(np.shape(N)[0]):
    ax.loglog(f,SxxTArray[i],label='N=%d'%N[i])
Sxx = (np.sin(np.pi*f*b)/(np.pi*f))**2.
ax.loglog(f,Sxx,'--',color='k',label='Sxx')
ax.set_xlabel(r'$f \ [Hz]$')
ax.set_ylabel(r'$S_{xx} \ [V^2Hz^{-1}]$')
ax.set_title('Spectral Density')
ax.legend()
````

Here we can see that $\lim_{T\rightarrow\infty}S_{xx}(f,T)$ (i.e. $N\rightarrow\infty$) does *not* converge to $S_{xx}(f)$. Looking at the plot it can be seen that the variance, or width of spead, of $S_{xx}(f,T)$ as $N\rightarrow\infty$ doesn't decreases to $S_{xx}(f)$.
