Using normally distributed random noise time series of length $N=2048$, show using a Monte Carlo analysis and by comparison to the theoretical probability distribution functions that:

   1. The raw spectral estimate is indeed distributed as $\chi^2_2$.

Hi All,  Assignment 5, Q1 discusses the distribution of spectral estimates.  These were discussed in Lecture 4, but because you have more blocks of data in Lecture 5 you have more degrees of freedom.

However, recall that it is the ratio of the spectral estimates to their expected value that follow the ($\frac{\chi^2_n}{n} $) distribution.  For the Gaussian random process the variance is 1 $V^2$, the Nyquist frequency is 0.5 Hz, if your sampling frequency is 1 Hz, so you should be able to calculate the expected value of the spectral estimate $E[G_{xx}]$ that you need to divide your Monte Carlo estimates by to get the proper distribution.

show using a Monte Carlo analysis and by comparison to the theoretical probability
distribution 

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
```

```python
N= 2048
T = 1024.
dt = T/N
nblocks = 8.
NFFT = int(N/nblocks)
wind = int(NFFT/2.)

#storage
GRaw = np.zeros((100,T))
fRaw = np.zeros((100,T))

Gblock = np.zeros((100,T/nblocks+1))
fblock = np.zeros((100,T/nblocks+1))
Gwind = np.zeros((100,T/nblocks+1))
fwind = np.zeros((100,T/nblocks+1))

#'n=numbers over runs':
t=np.arange(0,T,dt)
n = 100
for i in range(n):
    x=np.random.randn(N)
    # raw:
    X = dt*np.fft.fft(x)
    GRaw[i] = (2./T)*np.real(np.conj(X[:N/2])*X[:N/2])
    fRaw[i]=np.arange(N/2)/T
    #block
    Gblock[i],fblock[i] = mlab.psd(x,NFFT,Fs=1./dt)
    #wind
    Gwind[i],fwind[i] = mlab.psd(x,NFFT,Fs=1./dt,window=mlab.window_hanning,noverlap=wind)

print 'number of blocks:', nblocks
```

```python
#plot
fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(1,1,1) 
for ind in range(n):
    ax.loglog(fRaw[ind],GRaw[ind],'.',markersize=1, color='b') 
    ax.loglog(fblock[ind],Gblock[ind],'.',markersize=3, color='r') 
    ax.loglog(fwind[ind],Gwind[ind],'.',markersize=5, color='g') 
ax.set_xlabel(r'$f \ [Hz]$')
ax.set_ylabel(r'$G_{xx}(T,f) \ [(V)^2Hz^{-1}]$')
ax.set_title('Monte-Carlo of the power spectrum for spectral estimate (100 trials)')
```

I know this isn't a plot that was ask for. I did as more of reference for myself. It's just a good illustration of why the first hint is true.

Red = raw spectral estimate, block averaging = Blue, and 50% overlap Hanning windows = Green. 

```python
#flatten data for hist. (because HINT 1)
flat_GRaw = np.ravel(GRaw)
flat_Gblock = np.ravel(Gblock)
flat_Gwind = np.ravel(Gwind)

flat_Gwind1 = flat_Gwind #for Q1.2
nblocks1 = nblocks
```

```python
#plots
fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(1,1,1) 

#theoretical probability distribution:
tt = np.arange(0,100,1)

k_values = [2.,2.*nblocks]
colours = ['DarkRed','k']
for i,k in enumerate(k_values):
    y = stats.chi2.pdf(tt,k)
    #y1 = y/np.sum(y)
    ax.plot(tt,y,'--',label=r'$\chi^2_{%d}$'%k,color=colours[i])

#Monte Carlo:
ax.hist(flat_GRaw*2.,bins=100,range=[0,15],
        label='(normed) raw spectral estimate',normed=True,color='r', histtype='step',linewidth=2,alpha=0.5)
ax.hist(flat_Gblock*2.*nblocks,bins=100,
        label='(normed) block averaging',normed=True,color='b', histtype='step', alpha=0.5)
ax.hist(flat_Gwind*2.*nblocks,bins=100,
        label='(normed) 50% overlap Hanning windows',normed=True,color='g', histtype='step', alpha=0.5)
ax.set_xlim(0,40.)
ax.set_ylim(0,0.5)
ax.set_title('Comparison between theoretical to Monte Carlo probability distributions')
ax.set_xlabel(r'$G_{xx}(T,f) \ [(V)^2Hz^{-1}]$')
ax.set_ylabel('Distribution')
ax.legend()

print 'number of blocks:', nblocks
```

Here it can be seen theoretical  $\chi^2_{2}$ distribution fits the raw (normed) spectral estimate distribution for white noise. When a N block averaging is use to determine the spectral estimate this plot shows that it's distribution follows the theoretical  $\chi^2_{2N_{\text{block}}}$ distribution. However when a 50% overlap Hanning window is used to find the spectral estimate it's distribution close to block averaging, but has a bit lower variance. 

Note about scaling: I tried scaling $\frac{\chi^2_n}{n}$ to match the spectral estimate distribution however that made things worst. The only way I found that got $\chi^2_n$ distributions to match the spectral estimates was to scaling $G_{xx}$ by $2N_{\text{blocks}}$.

###### Show that as Nblocks is increased the variance drops (i.e. the distribution of the spectral estimates gets tighter):

```python
N= 2048
T = 1024.
dt = T/N

fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(1,1,1) 

nblocksArray = [8.,16.,24.]
for ind,nblocks in enumerate(nblocksArray):
    
    NFFT = int(N/nblocks)
    wind = int(NFFT/2.)

    #storage
    Gblock = np.zeros((100,T/nblocks+1))
    fblock = np.zeros((100,T/nblocks+1))
    Gwind = np.zeros((100,T/nblocks+1))
    fwind = np.zeros((100,T/nblocks+1))

    #'n=numbers over runs':
    t=np.arange(0,T,dt)
    n = 100
    for i in range(n):
        x=np.random.randn(N)
        #block
        Gblock[i],fblock[i] = mlab.psd(x,NFFT,Fs=1./dt)
        #wind
        Gwind[i],fwind[i] = mlab.psd(x,NFFT,Fs=1./dt,window=mlab.window_hanning,
                                     noverlap=wind)
    #flatten data for hist.
    flat_Gblock = np.ravel(Gblock)
    flat_Gwind = np.ravel(Gwind)
    
    #Monte Carlo:
    ax.hist(flat_Gblock,bins=100,
            label='(normed) block avg (nblock=%d)'%nblocks,normed=True,color='Blue', histtype='step', alpha=1.-6./nblocks)
    print 'The variance is:', np.std(flat_Gblock), 'for 50 percent block avg nblock=%d'%nblocks
    ax.hist(flat_Gwind,bins=100,
            label='(normed) Hanning wind (nblock=%d)'%nblocks,normed=True,color='DarkGreen', histtype='step', alpha=1.-6./nblocks)
    print 'The variance is:', np.std(flat_Gwind), 'for Hanning wind nblock=%d'%nblocks
ax.set_xlim(0,2.5)
#ax.set_ylim(0,0.1)
ax.set_title(r'Distribution of the spectral estimates as $N_{blocks}$ is increased')
ax.set_xlabel(r'$G_{xx}(T,f) \ [(V)^2Hz^{-1}]$')
ax.set_ylabel('Distribution')
ax.legend()
```

I'm sure there's better ways to display this. However it can be seen that "as Nblocks is increased the variance drops (i.e. the distribution of the spectral estimates gets tighter)" for both block averaging and 50% Hanning windowing.

## Q2: Spectra of a "red-noise" signal

<div style='background:#F0F0F0'> **1** Load in the Deep Cove hourly data, and compute the power spectra using some reasonable value for $N_{FFT}$, and comment on the effect of applying the Hanning window to not applying it to the spectral leakage.</div>

```python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
dc=dc[np.isfinite(dc)]
```

```python
def rawperio(x,fs):
    dt=1./fs
    N = len(x)
    T = N*dt
    X = dt*np.fft.fft(x)
    G = (2./T)*np.real(np.conj(X[:N/2])*X[:N/2])
    f= np.arange(N/2)/T
    return G[1:],f[1:]
```

```python
import matplotlib.mlab as mlab
dt = 3600.;
Gcanned,fcanned = mlab.psd(dc,NFFT=256,Fs=1./dt,window=mlab.window_hanning,noverlap=128)

G,f=rawperio(dc,1./dt)

fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(1,1,1)

ax.loglog(f,G,color='k',label='Raw', linewidth=0.1)
ax.loglog(fcanned,Gcanned,'b',label='50% Hanning windowed')

ax.annotate('Annual', xy=(1./3600/24/365, 1.e3),  xycoords='data',
                xytext=(1./3600/24/365, 1.e-2), textcoords='data',
                horizontalalignment='center',
                arrowprops=dict(arrowstyle="->")
                )
ax.annotate('Daily', xy=(1./3600/24, 1.e3),  xycoords='data',
                xytext=(1./3600/24, 1.e-2), textcoords='data',
                horizontalalignment='center',
                arrowprops=dict(arrowstyle="->")
                )

ax.set_xlabel(r'$f \ [Hz]$')
ax.set_ylabel(r'$G_{xx}(T,f) \ [(^oC)^2Hz^{-1}]$')
ax.set_title('The power spectrum for spectral estimate for Deep Cove')
ax.legend(loc=0)
```

Here it can be seen that the power spectra of the deep cove data simplified when a 50% Hanning window is applied. It has smoothed out a lot of the noise in the frequency. The frequency spike for the daily temperature us lot more apparent.  However there is a cost to do this. The lower end of the power spectrum has now been lost. The annual peak in frequency has now gone. There is increase  in slope at lower end of the spectra. However this just shows that the lost part of the power spectra does increase , but it's pretty limit information. 

If one is trying to determine a trend in higher frequency of a noisy power spectrum then 50% Hanning windowing works well. However if one were looking at the lower end spectra this would not be a good method.
